```python
import pandas as pd
import requests
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv", sep=";")
df.shape
```

```python
df.info()
```

# Добавление нового поля

```python
df["good"] = df["quality"].apply(lambda x: 1 if x >5 else 0)
df.head(20)
```

```python
df.describe()
```

```python
plt.style.available
```

# Влияние кислотности на качество вина

```python
plt.figure(figsize=(10,7))


plt.hist(df["fixed acidity"].loc[df["good"] == 0], alpha = 0.5 , label = "not good")
plt.hist(df["fixed acidity"].loc[df["good"] == 1], alpha = 0.5, label = "good")

plt.title("Значения кислотности и качество вина")
plt.xlabel("Кислотность")
plt.ylabel("Количество")

plt.legend()

plt.show()
```

# pH

```python
plt.figure(figsize=(10,7))


plt.hist(df["pH"].loc[df["good"] == 0], alpha = 0.5 , label = "not good")
plt.hist(df["pH"].loc[df["good"] == 1], alpha = 0.5, label = "good")

plt.title("Значения pH и качество вина")
plt.xlabel("pH")
plt.ylabel("Количество")

plt.legend()

plt.show()
```

```python
data1 = df.groupby("density")["good"].sum().reset_index()
data1
```

```python
plt.figure(figsize=(10,7))

plt.scatter(data1["density"],data1["good"])

plt.title("График разброса по плоности")
plt.xlabel("Плотность")
plt.ylabel("Количество")


plt.show()
```

# Качество вина по 3 химическим характеристикам

```python
data2 = df["volatile acidity"].loc[df["good"]==1]
data3 = df["citric acid"].loc[df["good"]==1]
data4 = df["residual sugar"].loc[df["good"]==1]
data4.head()
```

```python
plt.figure(figsize=(10,7))

plt.boxplot([data2,data3, data4], showfliers=False)

plt.title("Качество вина по химическим характеристикам")
plt.xlabel("Химические характеристики")
plt.ylabel("Количество")

plt.xticks([1,2,3],["volatile acidity","citric acid","residual sugar"])

plt.show()
```

# Максимальные значения химических характеристик для хорошего и плохого качества вина.

```python
data5 = df.groupby("good")[["chlorides","sulphates","alcohol","pH","density","fixed acidity"]].max().unstack("good").unstack("good").reset_index()
data6 = data5.rename(columns={0 : "no good", 1 : "good"})
data6
```

```python
plt.figure(figsize=(10,7))

n_ticks= np.arange(len(data6["good"]))
offset=0.2
w=0.4

plt.bar(n_ticks-offset,data6["good"],width=w,label="good")
plt.bar(n_ticks+offset,data6["no good"],width=w,label="no good")

plt.title("Максимальные значения химеческих харакеристик для хорошего и плохого качества вина")
plt.xlabel("Химические характеристики")
plt.ylabel("Максимальные числа")

plt.xticks(n_ticks,data6["index"],rotation=45)
plt.legend()

plt.show()
```

```python
plt.figure(figsize=(10,7))


plt.bar(data6["index"],data6["good"],label="good")
plt.bar(data6["index"],data6["no good"],bottom=data6["good"],label="no good")

plt.title("Максимальные значения химеческих харакеристик для хорошего и плохого качества вина")
plt.xlabel("Химические характеристики")
plt.ylabel("Максимальные числа")

plt.xticks(rotation=45)
plt.legend()

plt.show()
```

# Корреляция 

```python
data7 = df[["fixed acidity","volatile acidity","citric acid","residual sugar","chlorides","free sulfur dioxide","total sulfur dioxide","density","pH","sulphates","alcohol","quality"]].loc[df["good"]==1]
data7
```

```python
correlations = data7.corr()
correlations
```

```python
plt.figure(figsize=(10,7))
sns.heatmap(correlations,cmap="BuPu", annot=True)
plt.title("Корреляционная матрица")

plt.show()
```

```python
sns.jointplot(x= data7["quality"] , y= data7["alcohol"], data = data7, kind="reg")
```
